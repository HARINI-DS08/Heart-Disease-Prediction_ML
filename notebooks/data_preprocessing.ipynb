```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Heart disease data collection (1).xlsx", sheet_name="Sheet1")
print("Original shape:", df.shape)
df.head()
```

```python
#Understanding the data
df.info()
```

```python

df.describe()
```

```python
#Outlier removal
# Removeing biologically impossible values
df = df[(df["AGE"] >= 18) & (df["AGE"] <= 100)]
df = df[(df["HEIGHT (m)"] >= 120) & (df["HEIGHT (m)"] <= 230)]
df = df[(df["WEIGHT (kg)"] >= 30) & (df["WEIGHT (kg)"] <= 250)]
df = df[(df["TOTAL CHOLESTROL (mg/dL)"] >= 100) & (df["TOTAL CHOLESTROL (mg/dL)"] <= 450)]
df = df[(df["HDL (mg/Dl)"] >= 20) & (df["HDL (mg/Dl)"] <= 100)]
df = df[(df["LDL (mg/Dl)"] >= 50) & (df["LDL (mg/Dl)"] <= 250)]
df = df[(df["SYSTOLIC BLOOD PRESSURE (mmHg)"] >= 85) & (df["SYSTOLIC BLOOD PRESSURE (mmHg)"] <= 250)]
df = df[(df["DIASTOLIC BLOOD PRESSUE (mmHg)"] >= 45) & (df["DIASTOLIC BLOOD PRESSUE (mmHg)"] <= 150)]
df = df[(df["BMI"] >= 14) & (df["BMI"] <= 60)]
df = df[(df["PULSE PRESSURE (mmHg)"] >= 20) & (df["PULSE PRESSURE (mmHg)"] <= 100)]

df.to_excel("Heart_disease_data_cleaned (1) (1).xlsx", index=False)
files.download("Heart_disease_data_cleaned (1) (1).xlsx")
```
