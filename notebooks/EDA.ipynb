```python
# 1. Upload file
from google.colab import files
uploaded = files.upload()

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load dataset
df = pd.read_excel("Heart_disease_data_cleaned (1) (1).xlsx")

# Clean column names (remove spaces & make uppercase for consistency)
df.columns = df.columns.str.strip().str.upper()
print("Cleaned Columns:", df.columns.tolist())

# 3. Quick overview
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)

# 4. Summary statistics
print("\nSummary Stats:\n", df.describe())

# 5. Risk Score Distribution
sns.histplot(df['RISK SCORE'], bins=15, kde=True)
plt.title("Risk Score Distribution")
plt.show()

# 6. Risk Category Count
sns.countplot(x='RISK CATEGORY', data=df)
plt.title("Risk Category Distribution")
plt.show()

# 7. Age vs Risk Score
sns.scatterplot(x='AGE', y='RISK SCORE', hue='SEX', data=df)
plt.title("Age vs Risk Score")
plt.show()

# 8. BMI vs Risk Score
sns.scatterplot(x='BMI', y='RISK SCORE', hue='SMOKING', data=df)
plt.title("BMI vs Risk Score (Smoking Influence)")
plt.show()

# 9. Correlation Heatmap (only numeric columns)
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```
